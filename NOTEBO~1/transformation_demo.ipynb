# MuleSoft-like Data Transformation Demo

This notebook demonstrates the core functionality of the `DataTransformer` kit, mimicking MuleSoft's DataWeave logic in pure Python using `pandas` and `xmltodict`.

```python
import pandas as pd
import xmltodict
import json
import io

class DataTransformer:
    @staticmethod
    def json_to_csv(json_data):
        df = pd.DataFrame(json_data)
        return df.to_csv(index=False)

    @staticmethod
    def csv_to_json(csv_data):
        df = pd.read_csv(io.StringIO(csv_data))
        return df.to_dict(orient='records')

    @staticmethod
    def json_to_xml(json_data, root_name='root'):
        xml_dict = {root_name: json_data}
        return xmltodict.unparse(xml_dict, pretty=True)

    @staticmethod
    def xml_to_json(xml_data):
        return xmltodict.parse(xml_data)
```

## 1. JSON to CSV
Convert a list of items to CSV format.

```python
json_data = [
    {"id": 101, "name": "Laptop", "price": 1200},
    {"id": 102, "name": "Smartphone", "price": 800}
]

csv_out = DataTransformer.json_to_csv(json_data)
print("CSV Result:\n")
print(csv_out)
```

## 2. JSON to XML
Convert a dictionary to XML structure.

```python
single_item = {"id": 301, "name": "Headphones", "status": "In Stock"}
xml_out = DataTransformer.json_to_xml(single_item, root_name='Product')
print("XML Result:\n")
print(xml_out)
```
